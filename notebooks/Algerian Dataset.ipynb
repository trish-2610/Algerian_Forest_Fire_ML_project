## Data Cleaning 

```python
## importing libraries 
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
%matplotlib inline
```

```python
## loading data-set 
df = pd.read_csv("Algerian_dataset.xls",header=1)
```

```python
df
```

```python
df.info()
```

```python
## checking for null values 
df.isnull().sum()
```

```python
df[df.isnull().any(axis=1)]
```

```python
## making a new column region to divide dataset
df.loc[:122,"Region"] = 0 ## bejaia region
df.loc[122:,"Region"] = 1 
```

```python
df.info()
```

```python
## changing datatype to int
df["Region"] = df["Region"].astype(int)
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df = df.dropna().reset_index(drop=True)
df.head()
```

```python
df.isnull().sum()
```

```python
## now data has no null values 
```

```python
df.loc[[122]]
```

```python
## we need to remove this row 
df = df.drop(122).reset_index(drop=True)
df.loc[[122]]
```

```python
df.columns
```

```python
## renaming columns 
## fixing column spaces 
df.columns = df.columns.str.strip()
df.columns
```

```python
## Changing the data-type of columns 
df[["day","month","year","Temperature","RH","Ws"]] = df[["day","month","year","Temperature","RH","Ws"]].astype(int)
```

```python
df.info()
```

```python
objects = [features for features in df.columns if df[features].dtypes == "O"]
print(objects)
for i in objects:
    if i!= "Classes":
        df[i] = df[i].astype(float)
df.info()
```

```python
## df[["Rain","FFMC","DMC","DC","ISI","BUI","FWI"]] = df[["Rain","FFMC","DMC","DC","ISI","BUI","FWI"]].astype(float)
```

```python
df.describe()
```

```python
## saving the cleaned data-set into csv file 
df.to_csv("Algerian_cleaned.csv", index = False )
```

## Exploratory Data Analysis

```python
df_copy = df 
```

```python
df_copy.head() ## we have to predict FWI ( dependent / output feature )
```

```python
## No need of day , month , year column , so we are dropping it 
df_copy.drop(columns = ["day","month","year"] , axis = 1 , inplace = True )
```

```python
df_copy.head()
```

```python
df_copy["Classes"].value_counts()
## we have got multiple categories for "fire" and "not fire" because of spaces 
```

```python
## Encoding of categories in Classes column 
df_copy["Classes"] = np.where(df_copy["Classes"].str.contains("not fire"),0,1)
## np.where means wherever the str contains not fire make it 0 and where there is fire make it 1 
```

```python
df_copy["Classes"].value_counts()
```

## Data Visualization

```python
## FWI is our dependent/output features , rest all are our Independent/input features 
```

```python
## Density Plot for all features 
df_copy.hist(bins=30,figsize=(20,15)) ## we are using hist plot because most of the features are numerical 
plt.show()
```

```python
## plotting pie chart for Classes 
plt.figure(figsize=(4,4))
plt.pie(x = df_copy["Classes"].value_counts() , labels = ["fire","not fire"] ,\
        shadow = True ,autopct = "%1.1f", data = df_copy )
plt.title("Pie chart for Classes")
plt.show()
```

#### Correlation

```python
df = pd.read_csv("Algerian_cleaned.csv")
df["Classes"] = np.where(df["Classes"].str.contains("not fire"),"NOT FIRE","FIRE")
```

```python
## Monthly Fire Analysis 
df_temp = df[df["Region"]==0] ## Brajia Region 
plt.figure(figsize=(15,5))
sns.countplot(x="month",hue="Classes",data=df_temp,palette="rainbow")
plt.title("Fire analysis for Brajia Region")
plt.ylabel("Numnber of fire")
plt.show()
```

```python
## In month 8(August) there is most no. of fires 
## In month 6(June) there is most no. of Not fires 
```
